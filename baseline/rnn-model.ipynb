```python
import torch
import torch.nn as nn

# 定义输入数据
text = torch.randn(5, 3, 10)

# 定义LSTM模型
lstm = nn.LSTM(input_size=10, hidden_size=100, num_layers=2, bidirectional=True, batch_first=True)

# 进行前向计算
output, (h_n, c_n) = lstm(text)

# 打印输出
print("Output shape: ", output.shape)
print(output)
print("Output shape: ", h_n.shape)
print(h_n)
print("Output shape: ", c_n.shape)
print(c_n)
```
